````python
dfus = pd.read_csv('/content/data/rates_us.csv', header=0, sep=',')
dfus.head(20)


# create a pandas DataFrame with the following columns:
# Year
# Month
# Current_Rate (Interest rate for that month)
# Change_From_Last_Month (Difference in Current_Rate from the previous month)
# dfus = pd.DataFrame(columns=['Year', 'Month', 'Current_Rate', 'Change_From_Last_Month'])
# dfus
# dfus['Year'] == dfreadus['Date']
# dfus['Month'] == dfreadus['Date']
# dfus['Current_Rate'] == dfreadus['1 Mo']
# dfus['Change_From_Last_Month'] == dfreadus['1 Mo'] - dfreadus['1 Mo'].shift(1)

dfus['Date'] = pd.to_datetime(dfus['Date'])
dfus['Year'] = dfus['Date'].dt.year
dfus['Month'] = dfus['Date'].dt.month
dfus['Current_rate'] = dfus['1 Mo']
dfus['Change_From_Last_Month'] = dfus['1 Mo'].diff()
dfus = dfus.drop(['1 Mo'], axis=1)
dfus = dfus.drop(['Date'], axis=1)
dfus = dfus.drop(['2 Mo'], axis=1)
dfus = dfus.drop(['3 Mo'], axis=1)
dfus = dfus.drop(['4 Mo'], axis=1)
dfus = dfus.drop(['6 Mo'], axis=1)
dfus = dfus.drop(['1 Yr'], axis=1)
dfus = dfus.drop(['2 Yr'], axis=1)
dfus = dfus.drop(['3 Yr'], axis=1)
dfus = dfus.drop(['5 Yr'], axis=1)
dfus = dfus.drop(['7 Yr'], axis=1)
dfus = dfus.drop(['10 Yr'], axis=1)
dfus = dfus.drop(['20 Yr'], axis=1)
dfus = dfus.drop(['30 Yr'], axis=1)
#למחוק שורות שהשנה והחודש יהיו שוים
dfus = dfus.drop_duplicates(['Year', 'Month'])
dfus.head(20)
````

````python
# your code here
dfis = pd.read_csv('/content/data/rates_is.csv', header=0, sep=',')
dfis.head()
# create a pandas DataFrame with the following columns:
# Year
# Month
# Current_Rate (Interest rate for that month)
# Change_From_Last_Month (Difference in Current_Rate from the previous month)
dfis['TIME_PERIOD'] = pd.to_datetime(dfis['TIME_PERIOD'])
dfis['Year'] = dfis['TIME_PERIOD'].dt.year
dfis['Month'] = dfis['TIME_PERIOD'].dt.month
dfis['Current_rate'] = dfis['OBS_VALUE']
dfis['Change_From_Last_Month'] = dfis['OBS_VALUE'].diff()
dfis = dfis.drop(['OBS_VALUE'], axis=1)
dfis = dfis.drop(['TIME_PERIOD'], axis=1)
dfis = dfis.drop(['Series name'], axis=1)
dfis = dfis.drop(['FREQ'], axis=1)
dfis = dfis.drop(['SERIES_CODE'], axis=1)
dfis = dfis.drop(['Frequency'], axis=1)
dfis = dfis.drop(['IR_FV_TYPE'], axis=1)
dfis = dfis.drop(['Interest type'], axis=1)
dfis = dfis.drop(['Data source'], axis=1)
dfis = dfis.drop(['TIME_COLLECT'], axis=1)
dfis = dfis.drop(['Time period collection'], axis=1)
dfis = dfis.drop(['Unit of Measure'], axis=1)
dfis = dfis.drop(['DATA_SOURCE'], axis=1)
dfis = dfis.drop(['Confidentiality status'], axis=1)
dfis = dfis.drop(['PUB_WEBSITE'], axis=1)
dfis = dfis.drop(['Published on the website'], axis=1)
dfis = dfis.drop(['UNIT_MEASURE'], axis=1)
dfis = dfis.drop(['UNIT_MULT'], axis=1)
dfis = dfis.drop(['Unit multiplier'], axis=1)
dfis = dfis.drop(['RELEASE_STATUS'], axis=1)
dfis = dfis.drop(['RELEASE STATUS'], axis=1)
dfis = dfis.drop(['CONF_STATUS'], axis=1)
dfis = dfis.drop_duplicates(['Year', 'Month'])


dfis.head(30)


````
